### 3. Pruebas de eficiencia: (loops vs funciones vectorizadas)
Se puede apreciar que en la tabla que muestra los primeros 5 registros del conjunto de datos (_head(df1)_) se pueden apreciar dos columnas que llevan por nombre _description_ y _description__translated_. Si se desea realizar un estudio resumen del texto asociado a las descripciones de las personas que financian proyectos, el primer paso debería ser el obtener una columna única del texto en un solo idioma (inglés) para posteriormente realizar el trabajo de minería de datos textual.

Debido a que la columna _description__translated_ posee el texto traducido, el problema se aprecia en que si el texto original fue escrito en inglés, la nueva columna tiene valor nulo ya que todas las traducciones fueron adaptadas a inglés, por lo que una alternativa es verificar si es que la columna nueva tiene valor vacío y si es así traspasar el registro del idioma original y luego obtener la columna _description__translated_ completa.

#### 3.1 Estrategia ingenua (ciclos anidados)

La primera alternativa puede ser recorrer uno a uno cada registro de texto y revisar si es que existe registro nulo en _description__translated_, para luego llenarlo con el registro original (columna _description_).

```r
for(i in 1:dim(df1)[1]){
    if(df1$description_translated[i] == ""){df1$description_translated[i] = df1$description[i]}
}
```

```r
write.csv(df1,file = "data/df_backup.csv")
```

```r
df1 = fread("data/df_backup.csv",header = T,sep = ",")
head(df1)
```

```r
tail(df1)
```

```r
length(which(df1$description_translated==""))/dim(df1)[1]
```

Luego de 6 hrs 21 minutos, el loop no ha sido capaz de recorrer el conjunto de datos completo y realizar los cambios. Solo pudo con el 29.5$%$ de la cantidad total de registros.

#### 3.2 Estrategia eficiente (funciones vectorizadas)

Dado que son demasiados registros como para recorrer uno a uno en el conjunto de datos completo, se puede considerar una función a la cual se le puedan ingresar vectores y condiciones que haga el mismo procedimiento de comprobar si es que existen registros nulos para luego obtener la columna final con texto. 

```r
df1 = fread("data/loans.csv",header = T,sep = ",")
df1$description_translated = ifelse(df1$description_translated == "",df1$description,df1$description_translated)
head(df1)
```

```r
tail(df1)
```

```r
length(which(df1$description_translated==""))
```

Finalmente, existen 43285 registros sin texto.

### 4. Conclusiones:

* Si bien ambos procedimientos (3.1 y 3.2) son capaces de realizar la misma tarea, se ha comprobado una clara diferencia respecto a la eficiencia que involucra utilizar cada algoritmo.
* Utilizar loops, en contextos de una gran cantidad de registros, no es la manera más eficiente de realizar operaciones de búsqueda. Se puede ver en la siguiente figura que, además de requerir 6 hrs. y 21 minutos para ejecutar solo el 30$%$ inicial, el loop requiere la demanda de toda la memoria RAM (_Men_ en la esquina inferior izquierda) posible. Esto es una desventaja ya que no se puede disponer de memoria por un tiempo determinado y se corre el riesgo latente de que el computador se congele.

<img src="03.png">

* Efectivamente, utilizar funciones vectorizadas para recorrer vectores consume una menor cantidad de tiempo en comparación a loops y permite además realizar pruebas completas debido al poco tiempo que requiere la ejecución. En este caso, solo bastan 10.8 segundos para realizar todos los cambios.
* En este estudio no se han considerado loops anidados ya que el conjunto de datos tiene muchos registros por lo que es demasiada carga para el computador, pero se ha comprobado que las funciones vectorizadas son mucho más eficientes que un solo loop.
